## POST 방식

```python
url = 'https://search.naver.com/search.naver'
params = {
    'query': "디미고"
}
res = requests.get(url, params)
res.text
```

```python
r = requests.get("https://news.naver.com/main/read.naver?mode=LSD&mid=shm&sid1=105&oid=001&aid=0012592366", headers={
    'User-Agent': "asdf"
})

if r.ok: print(r.text[:1000])
else: print("Error")
```
