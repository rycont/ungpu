```python
from bs4 import BeautifulSoup as bs4
import requests
```

```python
r = requests.get("https://www.melon.com/chart/index.htm", headers={
    'User-Agent': "asdf"
})
```

```python
document = bs4(r.text)
```

```python
[link.text for link in document.select('[href*="javascript:melon.play.playSong"]')]
```

```python
song = [row.find("a").text for row in document.select(".ellipsis.rank01")]
```
