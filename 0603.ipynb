# 4. Bar Plot
* 막대 그래프로 여러 값들을 비교할 때 유용함
* 수치형 데이터 / 양적 데이터(Numerical Data): Y축
* 범주형 데이터(Categorical Data): X축

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
years = ['2018', '2019', '2020']
number = [100, 90, 80]

plt.bar(years, number)
plt.title("Births by year", size=20)
plt.xlabel('Yearn')
plt.ylabel('Births')
plt.show()
```

```python
# 수평 막대그래프 그리기
years = ['2018', '2019', '2020']
number = [100, 90, 80]

plt.barh(years, number, height=0.2)
plt.title("Births by year", size=20)
plt.xlabel('Yearn')
plt.ylabel('Births')
plt.show()
```

```python
# 연도와 성별로 출생자수 비교
male = [100, 90, 80]
femail = [90, 80, 70]
x = np.arange(3)

plt.bar(x - 0.1, male, width=0.2)
plt.bar(x + 0.1, femail, width=0.2)
plt.xticks(x, years)

plt.show()
```
