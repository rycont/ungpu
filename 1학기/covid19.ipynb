## 2.5. EDA(탐색적 데이터 분석)
### 2.5.1. 대한민국 코로나 데이터 분석

```python
df[df['국가'].str.contains('Korea')]
```

```python
koreaDf = df[df['국가'] == 'South Korea']
koreaDf
```

```python
sns.lineplot(x='날짜', y='확진자', data=koreaDf)
sns.lineplot(x='날짜', y='사망자', data=koreaDf)
sns.lineplot(x='날짜', y='회복자', data=koreaDf)
plt.legend(['Active', 'Dead', 'Recovered'])
plt.show()
```

## 2.5.2. 국가별 코로나 데이터 분석

```python
latestDf = df[df['날짜'] == df['날짜'].max()]
latestDf.head()
```

```python
sortedDf = latestDf.sort_values(by="확진자", ascending=False)
sortedDf.head(10)
```

```python
plt.figure(
    figsize=(10, 5)
)
sns.barplot(
    x='국가',
    y='확진자',
    data=sortedDf.head(10)
)
plt.xticks(rotation=45)
plt.show()
```

```python
worldWideDf = df.groupby('날짜')[['확진자', '사망자', '회복자']].sum().reset_index()
worldWideDf.head()
```

```python
sns.lineplot(x='날짜', y='확진자', data=worldWideDf)
sns.lineplot(x='날짜', y='회복자', data=worldWideDf)
sns.lineplot(x='날짜', y='사망자', data=worldWideDf)

plt.show()
```
