## 6.5. Box Plot (Box-Whisker Plot)
* 최소, 최대, 중앙값, 1 3분위수
* 이상치(Outlier) 감지

```python
sns.boxplot(x='year', y='passengers', data=df)
plt.show()
```

```python
df.loc[df.year == 1960].describe()
```

```python
sns.violinplot(x='year', y='passengers', data=df)
plt.show()
```

```python
sns.violinplot(x='year', y='passengers', data=df.loc[df.year == 1960])
plt.show()
```

```python
flights = df.pivot_table(index='month', columns='year', values='passengers')
```

```python
plt.figure(figsize=(20, 20))

sns.heatmap(
    data=flights,
    annot=True,
    fmt="d"
)
plt.show()
```

```python
sns.heatmap(
    data=tips.corr(),
    annot=True,
    fmt="f"
)
plt.show()
```
