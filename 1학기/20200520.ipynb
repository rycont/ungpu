```python
oneToTen = np.arange(1, 11)
plt.plot(oneToTen, oneToTen * 2)
plt.show()
```

```python
plt.plot(oneToTen, oneToTen * 2, marker="*", ms="18", ls="-.")
plt.plot(oneToTen * 2, oneToTen, ls=":", c="r", lw="12")
plt.show()
```

```python
plt.plot(oneToTen, oneToTen * 2, "o")
plt.plot(oneToTen, oneToTen * 2)
plt.show()
```

```python
plt.plot(oneToTen, oneToTen, 'o:r')
plt.plot(oneToTen, oneToTen * 2, '*-.b')
plt.show()
```

```python
plt.plot(oneToTen, oneToTen, 'o:r')
plt.figure(figsize=(5, 8))
# Figure :    
plt.plot(oneToTen, oneToTen, 'o:b')
plt.show()
```

```python
# Title, Label, Tick, Legend
plt.title("!", size=20)

plt.xlabel('', size=30)
plt.ylabel('', size=30)

plt.plot(oneToTen * 2, oneToTen)
plt.plot(oneToTen, oneToTen * 2)

plt.xticks()
plt.legend(['()', '()'])

plt.grid(
    ls=':'
) # axis ->  

plt.show()
```
